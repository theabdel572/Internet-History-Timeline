# Most Important Internet Events

- The data is in Spanish.

- The events are from 1937 to 1993.

```
import pandas as pd
data = pd.read_csv("events.txt", sep="|", engine='python', names=["Año", "Evento"])
data
```

```
import matplotlib.pyplot as plt
import numpy as np
# Making matplotlib work in the Jupyter Notebook.
%matplotlib inline

dates = data["Año"]

min_date = dates.min()
max_date = dates.max()

labels=[str(i).replace("\\n","\n") for i in data["Evento"]]

# Associating labels with the dates.
labels = ['{0}:\n{1}'.format(d, l) for l, d in zip (labels, dates)]

fig, ax = plt.subplots(figsize=(30, 8), constrained_layout=True)
timeline = ax.set_ylim(-2, 1.75)
timeline = ax.set_xlim(min_date, max_date)
timeline = ax.axhline(0, xmin=0.05, xmax=0.95, c='deeppink', zorder=1)
 
timeline = ax.scatter(dates, np.zeros(len(dates)), s=120, c='palevioletred', zorder=2)
timeline = ax.scatter(dates, np.zeros(len(dates)), s=30, c='darkmagenta', zorder=3)

label_offsets = np.zeros(len(dates))
label_offsets[::2] = 0.35
label_offsets[1::2] = -0.7
for i, (l, d) in enumerate(zip(labels, dates)):
    timeline = ax.text(d, label_offsets[i], l, ha='center', fontfamily='serif', fontweight='bold', color='royalblue',fontsize=9)

stems = np.zeros(len(dates))
stems[::2] = 0.3
stems[1::2] = -0.3
markerline, stemline, baseline = ax.stem(dates, stems)
timeline = plt.setp(markerline, marker=',', color='darkmagenta')
timeline = plt.setp(stemline, color='darkmagenta')

# Hiding some lines around the chart.
for spine in ["left", "top", "right", "bottom"]:
    timeline = ax.spines[spine].set_visible(False)

# Hiding tick labels.
timeline = ax.set_xticks([])
timeline = ax.set_yticks([])

timeline = ax.set_title('Eventos de internet', fontweight="bold", fontfamily='serif', fontsize=16, 
                 color='royalblue')
```
